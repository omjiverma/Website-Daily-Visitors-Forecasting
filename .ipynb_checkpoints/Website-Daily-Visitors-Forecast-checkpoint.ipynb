## Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly
import warnings
warnings.filterwarnings("ignore")
```

## Loading  Data

```python
raw_data = pd.read_csv('./daily-website-visitors.csv')
```

```python
raw_data.head()
```

### Data Preprocessing

```python
df = pd.DataFrame()
df['Date'] = pd.to_datetime(raw_data['Date'].values)
df['Visitors'] = raw_data['Page.Loads'].values
df.rename(columns={'Date':'ds','Visitors':'y'},inplace=True)
```

```python
df['y'] = df['y'].replace(',','', regex=True)
df['y'] = df['y'].astype(int)
```

### Data Exploration

```python
#Shape of Dataset
df.shape
```

```python
#Null Values
df.isna().sum()
```

```python
#Plot
plt.figure(figsize=(20,10))
plt.plot(df.ds,df.y)
plt.ylabel('Daily Website Visits')
plt.xlabel('Date')
plt.title('Daily Visitors',fontsize=(16))
```

### Fit Model

```python
m = Prophet()
m.fit(df);
```

```python
future = m.make_future_dataframe(periods=365)
forcast = m.predict(future)
```

### Seasonality

```python
plot_components_plotly(m, forcast)
```

### Forcast For Next 1 Year

```python
plot_plotly(m, forcast)
```

```python
plt.figure(figsize=(20,8))
plt.plot(forcast[-365:].ds,forcast[-365:].yhat);
plt.title('365 Days Forcast',fontsize=(16))
plt.ylabel('Visitors',fontsize=(12))
plt.xlabel('Date',fontsize=(12));
```
